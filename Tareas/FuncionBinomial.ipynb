Tarea de funcion binomial

```python
import numpy as np
import matplotlib.pyplot as plt
import cv2
import matplotlib.image as pyimg
```

```python
def create_filter(level):
    numerator = np.math.factorial(level)
    filter = []
    
    for i in range(level+1): 
        denominator = (np.math.factorial(i) * np.math.factorial(level-i))
        filter.append( numerator / denominator)
    
    filter = np.array([filter])
    filter = filter.T * filter

    return filter / filter.size
```

```python
filter = create_filter(2)
filter
```

```python
filter5 = create_filter(5)
filter5
```
