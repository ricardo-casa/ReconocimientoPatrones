```python
from PIL import Image, ImageFilter
import matplotlib.pyplot as plt
```

El radio representa el kernel del filtro, pero este se obtiene de valores gausianos

Imagen Original

```python
link = 'https://www.horshamconservatives.com/sites/www.horshamconservatives.com/files/2021-09/WIne%20and%20Cheese.jpg'

img = Image.open(r'wine_cheese.jpg')
plt.imshow(img)
plt.show()
```

Valor default de radio en GaussianBlur

```python
img = img.filter(ImageFilter.GaussianBlur)
plt.imshow(img)
plt.show()
```

Imagen a radio 5

```python
img = img.filter(ImageFilter.GaussianBlur(5))
plt.imshow(img)
plt.show()
```

Imagen a radio 10

```python
img = img.filter(ImageFilter.GaussianBlur(10))
plt.imshow(img)
plt.show()
```
